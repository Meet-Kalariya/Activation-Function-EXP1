# Sigmoid
Sigmoid takes a real value as input and outputs another value between 0 and 1.
Itâ€™s non-linear, continuously differentiable, monotonic, and has a **fixed output range**. The output of the activation function is always going to be in range (0,1) compared to (-inf, inf) of linear function.

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x) )

plt.figure(figsize=(10, 5))
plt.plot(x, y,)
plt.title("Sigmoid activation function")
plt.legend(['sigmoid function'])
plt.show()
```

# ReLU
ReLU stands for Rectified Linear Units. Itâ€™s not linear and provides the same benefits as Sigmoid but with **better performance**.
ReLu is less computationally expensive than tanh and sigmoid because it involves simpler mathematical operations. The range of ReLu is [0, inf).

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-10, 10, 1000)
y = np.maximum(0, x)

plt.figure(figsize=(10, 5))
plt.title("ReLU activation function")
plt.plot(x, y)
plt.legend(['Relu'])
plt.show()
```

#LeakyReLU
LeakyRelu is a variant of ReLU. Instead of being 0 when z<0, a leaky ReLU allows a small, non-zero, constant gradient Î± (alpha). Leaky ReLUs are one attempt to fix the **dying ReLU** problem by having a small negative slope (of 0.01, or so).

We can see below by changing the value of alpha the negative slope changes which results in different value for negative number.

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha*z, z)


z = np.linspace(-5, 5, 50)
fig = plt.figure(figsize=(12, 6))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
ax1.plot(z, leaky_relu(z, 0.05))
ax2.plot(z,leaky_relu(z,0.1))
ax1.set_title("alpha = 0.05")
ax2.set_title("alpha = 0.1")
ax1.grid(True)
ax2.grid(True)
fig.suptitle("Leaky ReLU activation function")
ax1.axis([-5, 5, -1, 5])
ax2.axis([-5, 5, -1, 5])
plt.show()
```

# TanH

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 1000)
y = (1 - np.exp(-2*x) ) / (1 + np.exp(-2*x) )

plt.figure(figsize=(10, 5))
plt.plot(x, y,)
plt.grid(True)
plt.title("Tanh activation function")
plt.legend(['Tanh function'])
plt.show()
```

# Exponential Linear Unit

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


def elu(z, alpha=0.01):
    return np.maximum(alpha*(np.exp(z)-1), z)

z = np.linspace(-10, 10, 10000)
fig = plt.figure(figsize=(12, 6))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
ax1.plot(z, elu(z, 0.05))
ax2.plot(z,elu(z,0.1))
ax1.set_title("alpha = 0.05")
ax2.set_title("alpha = 0.1")
minor_ticks = np.arange(-100, 101, 1)
major_ticks = np.arange(-100, 101, 5)
ax1.set_xticks(major_ticks)
ax1.set_xticks(minor_ticks, minor=True)
ax1.set_yticks(major_ticks)
ax1.set_yticks(minor_ticks, minor=True)
ax1.grid(which='minor', alpha=0.2)
ax1.grid(which='major', alpha=0.5)
ax2.set_xticks(major_ticks)
ax2.set_xticks(minor_ticks, minor=True)
ax2.set_yticks(major_ticks)
ax2.set_yticks(minor_ticks, minor=True)
ax2.grid(which='minor', alpha=0.2)
ax2.grid(which='major', alpha=0.5)
fig.suptitle("Exponential Linear Unit activation function")
ax1.axis([-7, 7, -2, 25])
ax2.axis([-7, 7, -2, 25])
plt.show()
```
