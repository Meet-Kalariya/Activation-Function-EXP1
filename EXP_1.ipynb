```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x) )

plt.figure(figsize=(10, 5))
plt.plot(x, y,)
plt.title("Sigmoid activation function")
plt.legend(['sigmoid function'])
plt.show()
```

```
x = np.linspace(-10, 10, 1000)
y = np.maximum(0, x)

plt.figure(figsize=(10, 5))
plt.title("ReLU activation function")
plt.plot(x, y)
plt.legend(['Relu'])
plt.show()
```

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


def leaky_relu(z, alpha=0.01):
    #print(np.maximum(alpha*z,z))
    return np.maximum(alpha*z, z)

z = np.linspace(-5, 5, 50)
#print(z)
plt.plot(z, leaky_relu(z, 0.05))
#props = dict(facecolor='black', shrink=0.1)
plt.grid(True)
#plt.annotate('Leak', xytext=(-3.5, 0.5), xy=(-5, -0.2), arrowprops=props, fontsize=14, ha="center")
plt.title("Leaky ReLU activation function")
plt.axis([-5, 5, -1, 5])
#plt.savefig("leaky_relu_plot")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5, 5, 1000)
y = (1 - np.exp(-2*x) ) / (1 + np.exp(-2*x) )

plt.figure(figsize=(10, 5))
plt.plot(x, y,)
plt.grid(True)
plt.title("Tanh activation function")
plt.legend(['Tanh function'])
plt.show()
```

```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


def elu(z, alpha=0.01):
    return np.maximum(alpha*(np.exp(z)-1), z)

z = np.linspace(-10, 10, 10000)
plt.plot(z, elu(z, 0.05))
plt.grid(True)
plt.title("Exponential Linear Unit activation function")
#plt.axis([-5, 5, -5, 5])
plt.show()
```
